# data 2
https://data.ny.gov/Energy-Environment/Gasoline-Retail-Prices-Weekly-Average-by-Region-Be/nqur-w4p7/data


```python
df2 = pd.read_csv('https://data.ny.gov/api/views/nqur-w4p7/rows.csv?accessType=DOWNLOAD&sorting=true')
df2.shape
```

```python
df2.head(100)
```

### Is this data tidy or not?

(Yes or no)

**Answer: No, the data is not tidy**

```python
#Converting the data to a tidy dataframe

new_df2=pd.melt(frame=df2,id_vars=['Date'],var_name='City',value_name='Average Price')
new_df2
```

```python
#Number of rows and columns in the new dataframe are:

print("The no. of columns in this new database are:", new_df2.shape[1])
print("The no. of rows in this new database are:", new_df2.shape[0])
```

```python
#The new column headers are:

list(new_df2.columns)
```

# data 3
https://catalog.data.gov/dataset/department-of-state-business-filings-beginning-1991

```python
df4 = pd.read_csv('https://data.ny.gov/api/views/m7i3-tv6j/rows.csv?accessType=DOWNLOAD')
df4.shape
```

```python
df4.head(10)
```

### Is this data tidy or not?

(Yes or no)

**Answer:** No, the data is not tidy.

*if the data is not tidy, then convert the data to a tidy dataframe*

```python
df4.info()
```

```python
df4.describe()
```

```python
#Converting the data to a tidy dataframe

new_df4=pd.melt(frame=df4,id_vars=['Full Certificate Code','Certificate Type Code','Certificate Type Description','Authority Type Code','Authority Type Description','Business Type Code','Business Type Description','Name Type Code','Name Type Description','Description','Year'],var_name='Month',value_name='Total')
new_df4
```

```python
#Number of rows and columns in the new dataframe are:

print("The no. of columns in this new database are:", new_df4.shape[1])
print("The no. of rows in this new database are:", new_df4.shape[0])
```

```python
#The new column headers are:

list(new_df4.columns)
```

# data 4
https://catalog.data.gov/dataset/new-york-city-leading-causes-of-death-ce97f


```python
df6 = pd.read_csv('https://data.cityofnewyork.us/api/views/jb7j-dtam/rows.csv?accessType=DOWNLOAD')
df6.shape
```

```python
df6
```

### Is this data tidy or not?

(Yes or no)

**Answer:** No, the data is not completely tidy

*if the data is not tidy, then convert the data to a tidy dataframe*

```python
df6.info()
```

```python
#Converting it into a tidy dataframe

df6["Sex"].replace("M", "Male", regex=True)
df6["Sex"].replace("F", "Female", regex=True)
df6.fillna(0)
df6.head(10)
```

```python
#Number of rows and columns in the new dataframe are:

print("The no. of columns in this new database are:", df6.shape[1])
print("The no. of rows in this new database are:", df6.shape[0])
```

```python
#The new column headers are:

list(df6.columns)
```
