### Loading the dataset into the jupyter notebook

----

```python
#The dataset has been obtained from the following website: https://www.worlddev.xyz/

df_1=pd.read_csv('https://raw.githubusercontent.com/ajitjadhav10/data690_ajit/main/individual_project_part_b/wdi_data.csv')
df_1.head(10)
```

```python
df_1.drop(columns=["Unnamed: 0"], inplace=True)
df_1
```

```python
#Only keeping the required countries

df_countries = df_1[df_1["Country Code"].isin(['AFG','BTN','IND','LKA','NPL','BGD','PAK'])]
df_countries
```

```python
#This is the dataset for women enrolled in secondary education for the selected south-asian countries

df_female_se = df_countries.query("indicator == 'SE.SEC.ENRL.FE.ZS'")
df_female_se
```

```python
df_pivot = df_female_se.pivot(index =['Year','Country Code','Country Name','Region','Income Group'],columns = 'indicator', values ='value').reset_index()
df_pivot.reset_index().head()
```

## Line graph - Percentage of women enrolled in higher level education:

---

```python
fig = px.line(
    df_pivot, 
    x="Year", 
    y="SE.SEC.ENRL.FE.ZS",
    labels={"SE.SEC.ENRL.FE.ZS":"Secondary education, pupils (% female)"},
    hover_name='Country Name',
    color='Country Name'
    )
fig.show()
```

```python
df_female_lf = df_countries.query("indicator == 'SL.TLF.CACT.FE.ZS'")
df_female_lf
```

```python
df_pivot_1 = df_female_lf.pivot(index =['Year','Country Code','Country Name','Region','Income Group'],columns = 'indicator', values ='value').reset_index()
df_pivot_1.reset_index().head()
```

## Line graph - Percentage of women in labor-force:

---

```python
fig = px.line(
    df_pivot_1, 
    x="Year", 
    y="SL.TLF.CACT.FE.ZS",
    labels={"SL.TLF.CACT.FE.ZS":"Female Labor force participation rate(% of female population ages 15+)"},
    hover_name='Country Name',
    color='Country Name',
    height=600
    )
fig.show()
```

```python
df_pivot_2 = df_countries.pivot(index =['Year','Country Code','Country Name','Region','Income Group'],columns = 'indicator', values ='value').reset_index()
df_pivot_2.reset_index().head()
```

## Scatter Plot - Percentage of women enrolled in higher-level education vs Percentage of women in labour-force

---

```python
df_pivot_2
fig = px.scatter(df_pivot_2, 
                x="SL.TLF.CACT.FE.ZS", 
                y="SE.SEC.ENRL.FE.ZS",
                title="'Percentage of women enrolled in higher-level education'vs Percentage of women in labour-force",
                labels={"SL.TLF.CACT.FE.ZS":"Labor force participation rate, female(% of female population ages 15+)(modeled ILO estimate)","SE.SEC.ENRL.FE.ZS":"Secondary education, pupils (% female)"},
                color="Country Name",
                hover_name='Year',
                height=700,
                size_max=7,
                size="SL.TLF.CACT.FE.ZS")
fig.update_traces(marker_size=40)
fig.show()
```

## Bar Chart - Female enrollment in secondary level education 2016


---

```python
df_pivot_3 = df_pivot.query("Year == '2016'")

fig = px.bar(df_pivot_3, x='Country Name', y='SE.SEC.ENRL.FE.ZS',title='Female enrollment in secondary level education - 2016'
             ,hover_data=['SE.SEC.ENRL.FE.ZS'], color='Country Name',
             labels={'SE.SEC.ENRL.FE.ZS':'2016 Secondary education, pupils(% females)'}, height=600).update_xaxes(categoryorder="total descending")
fig.show()
```

## Bar Chart - Female labor force participation 2016


---

```python
df_pivot_4 = df_pivot_1.query("Year == '2016'")

fig_1 = px.bar(df_pivot_4, 
               x='Country Name', 
               y='SL.TLF.CACT.FE.ZS',
               title='Female Labor force participation - 2016',
               labels={"SL.TLF.CACT.FE.ZS":"2016 Labor force participation rate, female(% of female population ages 15+)(modeled ILO estimate)"},
               hover_data=['SL.TLF.CACT.FE.ZS'], 
               color='Country Name',
               height=800).update_xaxes(categoryorder="total descending")
fig_1.show()
```
