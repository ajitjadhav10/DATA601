```python
# 1st Dataset: GDP per capita(current international $)

#This data is obtained from the world bank website
#Reading the GDP per capita data into a dataframe, skipping the first 4 rows because they are empty rows

df_1=pd.read_csv('https://raw.githubusercontent.com/ajitjadhav10/DATA601/main/Project_02/API_NY.GDP.PCAP.PP.CD_DS2_en_csv_v2_3158984.csv',skiprows=4)
df_1.head(2)
```

```python
#Tidying the data

#Making a single column for 'years' and the values of 'GDP per capita' 

df_1_tidy=pd.melt(df_1,
                  id_vars=['Country Name','Country Code','Indicator Name','Indicator Code','Unnamed: 65'],
                  var_name='Year',
                  value_name='value'
                  )
df_1_tidy.head(3)
```

```python
#Checking for null values

df_1_tidy.isnull().sum()
```

```python
#Removing null values: Keeping only the rows which have a numeric value for GDP per capita

df_1_1 = df_1_tidy[df_1_tidy['value'].notna()].reset_index()
df_1_1.head(3)
```

```python
#Dropping columns 'index' and 'Unnamed: 65' because they dont have any relevant data.
#This is the final clean dataset of GDP per capita

df_1_2=df_1_1.drop(columns=['Unnamed: 65','index'])
df_1_2.head(3)
```

```python
df_1_3= df_1_2.rename({'Country Name': 'Country_Name', 'Country Code': 'Country_Code', 'Indicator Name': 'Indicator_Name', 'Indicator Code': 'Indicator_Code'}, axis=1)
df_1_3.head()
```

```python
#Verifying if all null values have been removed.

df_1_3.isnull().sum()
```

```python
df_1_4=df_1_3.query("Country_Name in ['India','Pakistan','Sri Lanka','Bangladesh','Nepal','Afghanistan','Bhutan','South Asia','Brazil','Indonesia','Thailand','Philippines'] and Year in ['1990','1991','1992','1993','1994','1995','1996','1997','1998','1999','2000','2001','2002','2003','2004','2005','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020']")
```

```python
fig_1 = px.line(df_1_4,
               x='Year', 
               y='value',
               title='GDP per capita from 1990 to 2019',
               labels={"SL.TLF.CACT.FE.ZS":"2016 Labor force participation rate, female(% of female population ages 15+)(modeled ILO estimate)"},
               color='Country_Name',
               height=500)
fig_1.show()
```
